```python
#| export
'''kosha — FTS5 + vector search + call graph over your repo and installed packages.

Instantiate with `k = Kosha()` (uses `.kosha/` db in cwd by default).
Then call `doc(Kosha)` for the full API or jump straight to `k.context()`.

Key workflows:
  k.sync()                                         # index repo + env packages + call graph
  k.context("how does X work")                     # fan-out search with graph enrichment
  k.api_call_paths("fasthtml", "starlette")        # shortest call paths between two packages
  k.dep_stack(["myapp"], depth=2)                  # BFS dependency layers ordered by coupling
  k.public_api("fastcore")                         # public API entries for a package
  k.ni("fastcore.basics.merge")                    # node info: callers, callees, pagerank
'''

from kosha.core import Kosha
from pyskills.core import allow

__all__ = ['Kosha']
allow({Kosha: ...})
```

```python
from pyskills.core import xdir, __pytools__
import kosha.skill
```

```python
# Verify Kosha is accessible and allow is registered
assert 'Kosha' in xdir(kosha.skill), f"Kosha not in xdir: {xdir(kosha.skill)}"
assert Kosha in __pytools__, f"Kosha not registered in __pytools__"
print("skill module ok")
```

```python
from pyskills.core import list_pyskills
```

```python
# list_pyskills keys are ep.value (module path), so 'kosha.skill' not 'kosha'
skills = list_pyskills()
assert 'kosha.skill' in skills, f"'kosha.skill' not found in pyskills: {list(skills.keys())}"
print(f"kosha pyskill registered: {skills['kosha.skill'][:60]}")
```
