This module is the target of kosha's `pyskills` entry point
(`[project.entry-points.pyskills] kosha = "kosha.skill"` in `pyproject.toml`). A pyskills host
discovers it with `list_pyskills()` -- which reads the first paragraph of this module's docstring
via AST, without importing -- and loads it with `from kosha.skill import *` followed by `doc(kosha.skill)`.

The module is a curated re-export: `__all__` names exactly the surface an agent should use, so a
star import brings in the intended API and `doc()`/`xdir()` show only those symbols.

**When to use**: before reading a source file to learn an API, grepping for a function or pattern,
checking whether a dependency already does something, or deciding where new code goes.
**When to skip**: known config/lockfiles, pure file-structure tasks, or when the relevant kosha
results are already in context.

The workflow an agent follows after loading:

```python
k = Kosha()
k.status()                                  # always first: {files, packages, graph_nodes, stale_files, stale_pkgs}
k.sync(in_parallel=True)                    # only if what you need is stale; stale looks like missing
k.env_context('what you want', limit=8)     # packages only (faster)
k.context('what you want', graph=True)      # repo + packages + call-graph neighbours (default once a task spans modules)
k.ni('pkg.mod.fn')                          # callers / callees / registered peers
k.where_to_add('desc', limit=5)             # file:line insertion points for new code
k.public_api('pkg')                         # a package's real public API (__all__ + @patch)
```

`pkg_url(pkg)` returns the best docs/repo URL for a package -- feed it to a web fetcher for usage
examples. Filters combine with the query string: `package:fastcore`, `path:xtras`, `file:xtras*`,
`lang:py`, `type:FunctionDef`; bare package names auto-detect as `package:`.
